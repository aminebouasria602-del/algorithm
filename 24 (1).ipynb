```python
def check_and_update_mst(n, tree_edges, new_edge):
    new_v, new_w, new_cost = new_edge

    # Строим список смежности из рёбер текущего MST
    graph = [[] for _ in range(n)]
    for idx, (a, b, cost) in enumerate(tree_edges):
        graph[a].append((b, cost, idx))
        graph[b].append((a, cost, idx))

    # BFS для поиска пути от new_v до new_w в текущем MST
    visited = [False] * n
    parent = [-1] * n
    edge_to_parent = [-1] * n
    cost_to_parent = [0] * n

    queue = [new_v]
    visited[new_v] = True

    while queue:
        cur = queue.pop(0)
        if cur == new_w:
            break
        for neighbor, cost, edge_idx in graph[cur]:
            if not visited[neighbor]:
                visited[neighbor] = True
                parent[neighbor] = cur
                edge_to_parent[neighbor] = edge_idx
                cost_to_parent[neighbor] = cost
                queue.append(neighbor)

    # Идём назад от new_w до new_v, чтобы найти самое дорогое ребро на пути
    max_cost = -1
    max_edge_idx = -1
    cur = new_w

    while cur != new_v:
        if cost_to_parent[cur] > max_cost:
            max_cost = cost_to_parent[cur]
            max_edge_idx = edge_to_parent[cur]
        cur = parent[cur]

    # Если новое ребро не дешевле — оставляем текущий MST без изменений
    if new_cost >= max_cost:
        return True, tree_edges

    # Иначе — заменяем самое дорогое ребро на новое
    new_tree = [edge for idx, edge in enumerate(tree_edges) if idx != max_edge_idx]
    new_tree.append(new_edge)

    return False, new_tree


# Входные данные
n = 4
tree_edges = [(0, 1, 10), (1, 2, 5), (2, 3, 20)]
new_edge = (0, 3, 8)

is_same, result_tree = check_and_update_mst(n, tree_edges, new_edge)

print("MST не изменился" if is_same else "MST обновлён")
print("Рёбра:", result_tree)
```
