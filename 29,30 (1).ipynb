```python
def max_billboard_profit(positions, profits, min_distance=5):
    n = len(positions)
    
    billboards = sorted(zip(positions, profits))
    pos = [b[0] for b in billboards]
    rev = [b[1] for b in billboards]
    
    dp = rev[:]
    
    for i in range(n):
        for j in range(i):
            if pos[i] - pos[j] > min_distance:
                dp[i] = max(dp[i], dp[j] + rev[i])
    
    return max(dp)

# Example 
positions = [6, 7, 12, 14]
profits   = [5, 6,  5,  1]
print("Max profit:", max_billboard_profit(positions, profits))  # Expected: 10
```

```python
def max_independent_set(weights):
    n = len(weights)
    if n == 0: return 0
    if n == 1: return weights[0]
    
    dp = [0] * n
    dp[0] = weights[0]
    dp[1] = max(weights[0], weights[1])
    
    for i in range(2, n):
        dp[i] = max(dp[i-1], dp[i-2] + weights[i])
    
    return dp[n-1]

# Example 1 - 8 - 6 - 3 - 6
weights = [1, 8, 6, 3, 6]
print("Max weight:", max_independent_set(weights))  
```
