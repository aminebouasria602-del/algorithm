```python
def place_towers(houses, radius=4):
    houses = sorted(houses) 
    towers = []
    i = 0

    while i < len(houses):
        tower = houses[i] + radius
        towers.append(tower)

        while i < len(houses) and houses[i] <= tower + radius:
            i += 1

    return towers

# Пример
houses = [1, 5, 10, 15, 20, 23]
towers = place_towers(houses)
print(f"Вышки стоят на: {towers}")
print(f"Количество вышек: {len(towers)}")
```
