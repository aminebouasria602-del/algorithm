```python
#17
def min_cost_order(rates):
    # Сортируем лицензии по убыванию скорости роста
    licenses = sorted(enumerate(rates, 1), key=lambda x: -x[1])
    
    total = 0
    print("Оптимальный порядок покупки:")
    
    for month, (lic_num, r) in enumerate(licenses):
        # Стоимость лицензии в месяц t: 100 * r^t
        cost = 100 * (r ** month)
        total += cost
        print(f"  Месяц {month}: лицензия {lic_num} (r={r}) → ${cost:.2f}")
    
    print(f"\nОбщая стоимость: ${total:.2f}")

# Пример входных данных
rates = [1.5, 1.2, 1.8, 1.3]
min_cost_order(rates)
```

```python
#17(2)
def max_revenue_order(rates):
    # Сортируем по возрастанию коэффициента (самый быстро падающий — первым)
    equipment = sorted(enumerate(rates, 1), key=lambda x: x[1])
    
    total = 0
    print("Оптимальный порядок продажи:")
    
    for month, (eq_num, r) in enumerate(equipment):
        # Выручка от продажи оборудования в месяц t: 100 * r^t
        revenue = 100 * (r ** month)
        total += revenue
        print(f"  Месяц {month}: оборудование {eq_num} (r={r}) → ${revenue:.2f}")
    
    print(f"\nОбщая выручка: ${total:.2f}")

# Пример входных данных (r < 1)
rates = [0.9, 0.7, 0.5, 0.8]
max_revenue_order(rates)
```

```python
def is_subsequence(S_prime, S):
    i = 0
    for element in S:
        if i < len(S_prime) and element == S_prime[i]:
            i += 1
    return i == len(S_prime)

S = ["Amazon", "Yahoo", "eBay", "Yahoo", "Yahoo", "Oracle"]
S_prime = ["Yahoo", "eBay", "Yahoo", "Oracle"]

print(is_subsequence(S_prime, S))
```
