```python
def find(p, x):
    while p[x] != x:
        p[x] = p[p[x]]
        x = p[x]
    return x

def union(p, x, y):
    rx, ry = find(p, x), find(p, y)
    if rx == ry:
        return False
    p[ry] = rx
    return True

def solve(n, edges, k):
    p = list(range(n))
    tree = set()

    # build tree, anaza edges first
    for i, (u, v, t) in enumerate(edges):
        if t == 'A' and union(p, u, v):
            tree.add(i)
    for i, (u, v, t) in enumerate(edges):
        if t == 'B' and union(p, u, v):
            tree.add(i)

    if len(tree) != n - 1:
        return None

    have = sum(1 for i in tree if edges[i][2] == 'A')

    while have != k:
        remove_type = 'A' if have > k else 'B'
        add_type    = 'B' if have > k else 'A'
        done = False

        for out in list(tree):
            if edges[out][2] != remove_type:
                continue
            temp = tree - {out}
            p2 = list(range(n))
            for i in temp:
                union(p2, edges[i][0], edges[i][1])
            for inp, (u, v, t) in enumerate(edges):
                if t == add_type and inp not in tree:
                    if find(p2, u) != find(p2, v):
                        tree = temp | {inp}
                        have += -1 if remove_type == 'A' else 1
                        done = True
                        break
            if done:
                break

        if not done:
            return None

    return list(tree)


# example
edges = [(0,1,'A'),(1,2,'B'),(2,3,'A'),(0,3,'B'),(1,3,'A')]
result = solve(4, edges, k=2)
if result is None:
    print("Impossible")
else:
    for i in result:
        u, v, t = edges[i]
        print(f"{u}--{v} ({'Anaza' if t=='A' else 'Bani Yas'})")
```

```python
def solve(prices):
    min_price, min_day = prices[0], 0
    best_profit, buy, sell = 0, -1, -1

    for i in range(1, len(prices)):
        if prices[i] - min_price > best_profit:
            best_profit = prices[i] - min_price
            buy, sell = min_day, i
        if prices[i] < min_price:
            min_price, min_day = prices[i], i

    if best_profit == 0:
        print("No profit possible")
    else:
        print(f"Buy day {buy+1}, sell day {sell+1}, profit ${best_profit}")


solve([9, 1, 5])
```
