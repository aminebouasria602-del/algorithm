```python
#30 A
def greedy_max_weight(weights):
    """Жадный алгоритм: всегда берём узел с максимальным весом"""
    n = len(weights)
    available = list(range(n))   # доступные узлы
    selected = []
    w = weights[:]

    while available:
        # Найти узел с максимальным весом среди доступных
        best = max(available, key=lambda i: w[i])
        selected.append(best)

        # Удалить best и его соседей
        to_remove = [best]
        if best - 1 in available: to_remove.append(best - 1)
        if best + 1 in available: to_remove.append(best + 1)

        for node in to_remove:
            available.remove(node)

    total = sum(weights[i] for i in selected)
    return selected, total

# --- Контрпример ---
weights = [6, 10, 6]
#  граф:  6 — 10 — 6

nodes, profit = greedy_max_weight(weights)
print("Жадный выбрал узлы:", nodes, "→ прибыль:", profit)
# Выберет узел 1 (вес 10) → прибыль = 10

# Оптимальное решение: взять узлы 0 и 2 (6+6=12)
print("Оптимально: узлы [0, 2] → прибыль:", weights[0] + weights[2])
```

```python
#30 B
def s1_s2_algorithm(weights):
    """
    S1 = узлы с нечётными индексами (1, 3, 5, ...)
    S2 = узлы с чётными индексами  (2, 4, 6, ...)
    Вернуть тот, у кого больший суммарный вес
    """
    n = len(weights)

    # 1-индексация как в задаче
    s1 = [i for i in range(n) if (i + 1) % 2 == 1]  # нечётные: 0,2,4...
    s2 = [i for i in range(n) if (i + 1) % 2 == 0]  # чётные:   1,3,5...

    sum_s1 = sum(weights[i] for i in s1)
    sum_s2 = sum(weights[i] for i in s2)

    if sum_s1 >= sum_s2:
        return s1, sum_s1
    else:
        return s2, sum_s2


weights = [10, 1, 1, 10, 1]

nodes, profit = s1_s2_algorithm(weights)
print("S1/S2 выбрал узлы:", nodes, "→ прибыль:", profit)

print("Оптимально: узлы [0, 3] → прибыль:", weights[0] + weights[3])
```

```python
#30 c
def max_independent_set_dp(weights):
    """
    Правильный алгоритм через динамическое программирование
    dp[i] = максимальный вес независимого множества на первых i узлах
    """
    n = len(weights)
    if n == 0: return [], 0
    if n == 1: return [0], weights[0]

    dp = [0] * n
    dp[0] = weights[0]
    dp[1] = max(weights[0], weights[1])

    for i in range(2, n):
        dp[i] = max(dp[i-1], dp[i-2] + weights[i])

    # Восстановление: какие узлы выбрали?
    selected = []
    i = n - 1
    while i >= 0:
        if i == 0 or dp[i] != dp[i-1]:
            selected.append(i)
            i -= 2
        else:
            i -= 1

    selected.reverse()
    return selected, dp[n-1]

# Тест на контрпримере из 30(b)
weights = [10, 1, 1, 10, 1]
nodes, profit = max_independent_set_dp(weights)
print("ДП выбрал узлы:", nodes, "→ прибыль:", profit)  

# Тест из оригинальной задачи
weights2 = [1, 8, 6, 3, 6]
nodes2, profit2 = max_independent_set_dp(weights2)
print("ДП выбрал узлы:", nodes2, "→ прибыль:", profit2) 
```
