```python
#  (имя, плавание, велосипед, бег) 
participants = [
    ("Алиса", 3, 10, 5),
    ("Боб",   5, 4,  8),
    ("Кэрол", 2, 7,  3),
    ("Дэйв",  4, 12, 6),
    ("Ева",   6, 5,  9),
]

#  сортируем по убыванию (велосипед + бег)
# тот, кто дольше едет и бежит — плывёт первым
participants.sort(key=lambda p: p[2] + p[3], reverse=True)

#  симулируем гонку
print(f"{'Место':<6} {'Имя':<8} {'Начало плавания':<18} {'Время финиша'}")
print("-" * 45)

pool_free = 0  # когда бассейн освобождается

for i, (name, swim, bike, run) in enumerate(participants, 1):
    swim_end = pool_free + swim   # когда участник выходит из бассейна
    finish   = swim_end + bike + run  # когда участник финишрует

    print(f"{i:<6} {name:<8} {pool_free:<18} {finish}")

    pool_free = swim_end  # следующий участник ждёт освобождения бассейна

print("-" * 45)
print(f"Общее время: {finish} минут")
```
