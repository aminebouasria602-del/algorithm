```python

# Each participant: (name, swim, bike, run)
participants = [
    ("Alice", 3, 10, 5),
    ("Bob",   5, 4,  8),
    ("Carol", 2, 7,  3),
    ("Dave",  4, 12, 6),
    ("Eve",   6, 5,  9),
]

# Step 1: sort by (bike + run) descending
participants.sort(key=lambda p: p[2] + p[3], reverse=True)

# Step 2: simulate the race
print(f"{'Order':<6} {'Name':<8} {'Swims at':<10} {'Finishes at'}")
print("-" * 40)

pool_free = 0  # when the pool becomes free

for i, (name, swim, bike, run) in enumerate(participants, 1):
    swim_end = pool_free + swim
    finish   = swim_end + bike + run
    
    print(f"{i:<6} {name:<8} {pool_free:<10} {finish}")
    
    pool_free = swim_end  # next person can only start after this one exits

print("-" * 40)
print(f"Total time: {finish}")
```
